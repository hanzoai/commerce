```python
import pandas as pd
df = pd.read_csv('_export/stoned-order-crowdstart-us-2017-01-26.csv', infer_datetime_format=True, usecols=['CreatedAt', 'Total'], parse_dates=['CreatedAt'])
```

```python
df.set_index(df['CreatedAt'], inplace=True)
```

```python
df.index = df.index.tz_localize('UTC').tz_convert('US/Central')
```

```python
orders_by_day = df['Total'].resample('D', how='sum')
```

```python
orders_by_day.to_csv("orders_daily.csv")
```
